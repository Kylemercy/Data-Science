## how to handle missig data

```python
kd = pd.read_csv('weather_data2.csv')
kd
```

```python
#to convert day column to date
kd = pd.read_csv('weather_data2.csv',parse_dates=['day'],index_col='day')
# to set day column as index
kd
```

```python
#to replace na values with 0
new_kd = kd.fillna(0)
new_kd
```

```python
# to specify different values for different column using fillna
new_kd = kd.fillna({'temperature':0,
                   'windspeed':0,
                   'event':'no event'})
new_kd
```

```python
# to carryover our previous data to fillin for missing values
#ffill = foward value
new_kd = kd.fillna(method='ffill')
new_kd
```

```python
# we also have backfill
new_kd = kd.fillna(method='bfill')
new_kd
```

```python
#this copies values horizontally for the ffill insted of vertically only works with bfill
new_kd = kd.fillna(method='bfill',axis='columns')
new_kd
```

```python
new_kd = kd.fillna(method='ffill',limit=1)
#to limit number of times the value is copied
new_kd
```

```python
new_kd = kd.interpolate()
new_kd
#using interpolate to guess for our nan values only for integers
```

```python
new_kd = kd.interpolate(method='time')
new_kd
# to interpolate our data using time
```

### AbstractHolidayCalendar

```python
# to define your own country holiday 
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class myCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('My Birth Day', month=4, day=12),#, observance=nearest_workday),
    ]
    
my_bday = CustomBusinessDay(calendar=myCalendar())
```

```python
#note this wont print weekends
pd.date_range(start='6/1/2001',end='6/30/2001',freq=my_bday)
#from this we see that our date sries lacks our bday date
```

```python
#in a case where the holiday is on weekend in other to oberseve the holiday on a workday we do this
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class myCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('My Birth Day', month=6, day=17 , observance=nearest_workday),
    ]
    
my_bday2= CustomBusinessDay(calendar=myCalendar())
```

```python
pd.date_range(start='6/1/2001',end='6/30/2001',freq=my_bday2)
# here the holiday is on a saturday , therefore the holiday is shifted to the nereast workday friday
```

```python
#for countries like eygypt their weekened is on friday and saturday sunday is a workday to create a 
#calender for this country, we create a custom weeday
b = CustomBusinessDay(weekmask='Sun Mon Tue Wed Thu')#fri and sat is weekend
pd.date_range(start='6/1/2023',end='6/30/2023',freq=b)
```

```python
#to specify a holiday
b = CustomBusinessDay(weekmask='Sun Mon Tue Wed Thu',holidays=['2023-06-27'])#fri and sat is weekend
pd.date_range(start='6/1/2023',end='6/30/2023',freq=b)
```
