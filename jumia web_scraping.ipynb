```python
Url = 'https://books.toscrape.com/'
response = requests.get(Url)
```

```python

Soup = BeautifulSoup(response.content,'html.parser')
books = Soup.prettify()
```

```python
books
```

```python
ol= Soup.find('ol')
#o find the firstinstance of ol
```

```python
ol
```

```python
#to find product name
product = ol.find_all('article',class_ = 'product_pod')
```

```python
for content in product:
    image = content.find('img')
    title = image.attrs['alt']
    print(title)
```

```python
#to get the stars of each book
for content in product:
    image = content.find('img')
    title = image.attrs['alt']
    star = content.find('p')
    star = star['class'][1]
    print(star)
```

```python
#to get the price 
for content in product:
    image = content.find('img')
    title = image.attrs['alt']
    price = content.find('p',class_ = 'price_color').get_text()
    price = float(price[1:])

    print(price)
```

```python
#to create a function that allows us move from first webpage to the last webpage 
Books = []
for i in range(1,51):
    Url = f'https://books.toscrape.com/catalogue/page-{i}.html'
    response = requests.get(Url)
    Soup = BeautifulSoup(response.content,'html.parser')
    ol= Soup.find('ol')
    product = ol.find_all('article',class_ = 'product_pod')
   
    for content in product:
        image = content.find('img')
        title = image.attrs['alt']
        star = content.find('p')
        star = star['class'][1]
        price = content.find('p',class_ = 'price_color').get_text()
        price = float(price[1:])
        Books.append([title,price,star])
```

```python
#to put our data into a dataframe
df1 = pd.DataFrame(Books,columns=['Title','Price','Stars'])
df1
```

```python
#to convert it to a csv file 
df1.to_csv('Books.csv')
```
