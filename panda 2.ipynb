### DATETIME 

```python
from datetime import datetime
d_parser = lambda x:datetime.strptime(x,'%Y-%m-%d %I-%p')
#%I reps 12 hour clock,%P reps pm or am
#to convert a string to dtetime we use striptime
cf = pd.read_csv('ETH_1h.csv',parse_dates=['Date'],date_parser = d_parser)
#this reads our date in form of datetime so we can use the datetimemethod on our date
```

```python
cf.head(10)
```

```python
cf.shape
```

```python
cf.loc[0,'Date']
```

```python
#to convert our data above to datetime
cf['Date'] = pd.to_datetime(cf['Date'],format='%Y-%m-%d %I-%p')
cf['Date']
```

```python
cf.loc[0,'Date'].day_name()
```

```python
cf['Date'].dt.day_name()
```

```python
cf['DayOfWeek'] = cf['Date'].dt.day_name()
```

```python
cf
```

```python
cf['Date'].min()
#min date
```

```python
cf['Date'].max()
#highest date
```
