# Bar Chart

```python
df2 = pd.read_csv('mat_data.csv')
df2
```

```python
df2.set_index('Responder_id',inplace=True)
df2
```

```python
from collections import Counter


language_count = Counter()
for row in df2['LanguagesWorkedWith']:
    language_count.update(row.split(';'))
print(language_count)
```

```python
#to print most common 15 languages
print(language_count.most_common(15))
```

```python
#to plot this data
language =[]
popularity = [ ]
for item in language_count.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
print(language)
print(popularity)
```

```python
#language, popularity = zip(*language_count.most_common(15)) 
#print(language) 
#print(popularity)
#using zip function
```

```python
plt.bar(language,popularity)


plt.title('Most Popular Languages',fontdict={'fontweight':'bold','fontsize':14})
plt.xlabel('Programming languages')
plt.ylabel('Number of people who use it')
plt.xticks(rotation=65)
# we use xticks to rotate the x_axis values to prevent overlapping of the values

plt.show()
```

```python
#alternatively we use a horizontal bar chart note we have to switch the x and  y label 
plt.barh(language,popularity,color = 'green')


plt.title('Most Popular Languages',fontdict={'fontweight':'bold','fontsize':14})
plt.ylabel('Programming languages')
plt.xlabel('Number of people who use it')
plt.show()
```

```python
#to have the most popular language on top
language.reverse()
popularity.reverse()
plt.barh(language,popularity,color = 'purple')


plt.title('Most Popular Languages',fontdict={'fontweight':'bold','fontsize':14})
plt.ylabel('Programming languages')
plt.xlabel('Number of people who use it')
```

# PIE CHART

```python
language1 =[]
popularity1 = [ ]
for item in language_count.most_common(7):
    language1.append(item[0])
    popularity1.append(item[1])
print(language1)
print(popularity1)
```

## FILLING PLOT

```python
data = pd.read_csv('fill_data.csv')
data.head(2)
```

```python
plt.figure(figsize=(8,5))
plt.style.use('default')
age = data['Age']
dev_salary = data['All_Devs']
py_salary = data['Python']
jav_salary = data['JavaScript']


plt.plot(age,dev_salary,'r--',label = 'all dev')
plt.plot(age,py_salary,'b',label = 'python')


plt.xlabel('Ages')
plt.ylabel('Median salary(USD)')
plt.fill_between(age,py_salary,alpha= 0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Days')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
#to plot between the python salary and over all median salary
plt.figure(figsize=(8,5))
plt.style.use('default')
age = data['Age']
dev_salary = data['All_Devs']
py_salary = data['Python']
jav_salary = data['JavaScript']


plt.plot(age,dev_salary,'r--',label = 'all dev')
plt.plot(age,py_salary,'b',label = 'python')


plt.xlabel('Ages')
plt.ylabel('Median salary(USD)')
overall_median = 57287
plt.fill_between(age,py_salary,overall_median,alpha= 0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Days')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
#to fill base on certain conditions


#plt.style.use('default')
age = data['Age']
dev_salary = data['All_Devs']
py_salary = data['Python']
jav_salary = data['JavaScript']




plt.plot(age, dev_salary, 'r--', label='all dev')
plt.plot(age, py_salary, 'b', label='python')

plt.xlabel('Ages')
plt.ylabel('Median salary(USD)')
overall_median = 57287
plt.fill_between(age, py_salary, overall_median, where=(py_salary > overall_median), interpolate=True, alpha=0.25, label= 'Above Avg')
plt.fill_between(age, py_salary, overall_median, where=(py_salary < overall_median), interpolate=True, alpha=0.25,label='Below Avg')

plt.title('Median Salary (USD) by Age')

plt.tight_layout()
plt.legend()
plt.show()
#values below the median value are colored red those  above are colored blue
```
