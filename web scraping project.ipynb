```python
#creating a function to grab the title and movie link for different movies
def get_content_value(row_data):
    if row_data.find('li'):
        return [li.get_text(' ',strip=True).replace('\xa0',' ') for li in row_data.find_all('li')]
    else:
        return row_data.get_text(' ',strip=True).replace('\xa0',' ')
    ## clean up the refrences[remove [1][2]]
def clean_tags(soup):
    for tag in soup.find_all('sup'):
        tag.decompose()
def get_info_box(url):
    r = requests.get(url)
    soup = bs(r.content)
    info_box = soup.find(class_= "infobox vevent")
    info_rows =info_box.find_all('tr')



    movie_info = {}
    #this enumerate all over the movies an append into the dictionary 
    for idx,row in  enumerate(info_rows):
        if idx == 0:
            movie_info['title'] = row.find('th').get_text(' ',strip=True)
        elif idx == 1:
            continue
        else:
            content_key = row.find('th').get_text(' ',strip=True)
            content_value = get_content_value(row.find('td'))
            movie_info[content_key] = content_value
    return movie_info
```

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films')
soup = bs(r.content)
movies2 = soup.select('.wikitable.sortable i a')
base_path = 'https://en.wikipedia.org/'

movie_info_list =[]
#this allows us to get a link element inside an italiced sentence
for idx,movie in  enumerate(movies2):
       try:
        relative_path = movie['href']
        full_path = base_path + relative_path 
        title = movie['title']
        #to get info box of all movie
        movie_info_list.append(get_info_box(full_path))
       
       except Exception as e:
            print(movie.get_text())
            print(e)
```

### to save the above data

```python
import json
def save_data(title,data):
    with open( title,'w',encoding = 'utf-8') as f:
        json.dump(data,f,ensure_ascii = False,indent = 2)
        
```

```python
#to load the saved data
def load_data(title):
    with open(title,encoding = 'utf-8') as f:
        return json.load(f)
```

```python
save_data('disney_movie_data.json',movie_info_list )
```
