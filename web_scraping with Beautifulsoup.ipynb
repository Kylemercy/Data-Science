## WEB SCRAPING THE 100 BEST MOVIES ON AMAZON PRIME VIDEO (JUNE 2024)

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
try:
    url = requests.get("https://editorial.rottentomatoes.com/guide/best-movies-on-amazon-prime-right-now/")
    url.raise_for_status()
    soup = bs(url.text,"html.parser")
    print(soup.prettify())
except Exception as e:
    print(e)
```

```python
movie_name = soup.select("div.article_movie_title a")
movie_name
```

```python
movie_names= [item.get_text() for item in movie_name]
movie_names
```

```python
#get ratings
movie_rating = soup.find_all(class_ = "tMeterScore")

movie_star= [item.get_text() for item in movie_rating]
movie_star
```

```python
#characters
movie_cast = soup.find_all("div",attrs={"class":"info cast"})
cast = [item.get_text().strip("\n")[10:] for item in movie_cast]
cast
```

```python
#director
movie_director = soup.find_all("div",attrs={"class":"info director"})
director = [item.get_text().strip("\n")[12:] for item in movie_director]
director
```

```python
#to convert this to a dataframe using pandas

top_rated = []

for i in range(len(movie_names)):
    movie_info = {
        'Movie_title': movie_names[i],
        'director': director[i],
        'cast': cast[i],
        'rating': movie_star[i]
    }
    top_rated.append(movie_info)

print(top_rated)
```

```python
#to convert this to a dataframe
df= pd.DataFrame.from_dict(top_rated)
df
```

```python
#to set indext to movie tite
new_idx = df.set_index("Movie_title")
new_idx
```

```python
#to perform pandas operations
data = df[['Movie_title','cast']]
data
```

## WEBSCRAPING 25 MOST POPULAR TV SHOWS RIGHT NOW: WHAT TO WATCH FROM ROTTEN TOMATOES

```python
try:
    url2 = requests.get("https://editorial.rottentomatoes.com/guide/popular-tv-shows/")
    url2.raise_for_status()
    soup2 = bs(url2.text,"html.parser")
  
except Exception as e:
    print(e)
```

```python
movies = soup2.find_all("div", attrs={"class": "row countdown-item"})
# to get movie rank, name, rating, cast, director
tomatoes = [ ]
for movie in movies:
    Rank = movie.find("div", class_="countdown-index").get_text()
    movie_title =movie.find("div",class_="article_movie_title").a.get_text()
    rating_element = movie.find("div", class_="article_movie_title").find_all(class_ = "tMeterScore")
    Rating_element = str([element.get_text() for element in  rating_element]).strip("[],'' ")
    cast_elements = movie.find("div", class_="info cast").find_all("a")
    Cast = str([element.get_text() for element in cast_elements])
    director_element = movie.find("div", class_= "info director").find_all("a")
    Director = str([element.get_text() for element in  director_element]) 
    tomatoes.append([Rank,movie_title,Rating_element,Cast, Director])
print(tomatoes)
```

```python
#to add this to a pandadas dataframe
df2 = pd.DataFrame(tomatoes,columns=["Rank","Title","Rating","Cast","Director"])
df2
```

```python
df2 = df2.set_index("Rank")
df2 
```
